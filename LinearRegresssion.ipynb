```python
%matplotlib inline
import numpy as np
from pylab import *
age = np.random.normal(40.0, 10.0, 1000)
height = 200 - (age + np.random.normal(20, 4, 1000))
```

```python
hist(age)
```

```python
hist(height)
```

```python
scatter(age, height)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(age, height)
print(r_value**2)
```

```python
import matplotlib.pyplot as plt
def predict(x):
    return slope * x + intercept
fitLine = predict(age)
plt.scatter(age, fitLine, c='r')
plt.show()
```
